# Estacionariedade

Muitos modelos tradicionais de séries temporais estatísticas dependem de que a série seja estacionária. De modo geral, uma série temporal estacionária é aquela que possui propriedades estatísticas relativamente estáveis ao longo do tempo, especialmente no que diz respeito à média e à variância. Isso parece relativamente simples.

As características de uma série temporal estacionária incluem:

1. **Média constante**: A média da série temporal permanece estável ao longo do tempo. Não há tendência de aumento ou diminuição contínua.

2. **Variância constante**: A variância, ou dispersão dos dados em torno da média, permanece constante ao longo do tempo. Ou seja, a amplitude entre os valores altos e baixos da série não muda.

3. **Covariância constante**: A autocorrelação da série temporal depende apenas da diferença de tempo entre dois pontos, e não do tempo em que as observações foram feitas. Isso significa que a relação entre valores em diferentes momentos no tempo permanece a mesma.

4. **Ausência de sazonalidade ou padrões cíclicos**: Uma série estacionária não deve apresentar ciclos ou padrões sazonais previsíveis que mudem com o tempo.

A estacionariedade é importante na prática por várias razões. Primeiro, muitos modelos, como os modelos tradicionais com forças bem conhecidas e modelos estatísticos, assumem que o processo é estacionário.

Um ponto mais amplo é que um modelo de uma série temporal que não seja estacionária terá sua precisão afetada à medida que as métricas da série variam. Ou seja, se você está tentando usar um modelo para estimar a média de uma série temporal cuja média e variância não são estacionárias, o viés e o erro do modelo vão variar ao longo do tempo, o que pode comprometer o valor do seu modelo.

Muitas vezes, uma série temporal pode ser tornada suficientemente estacionária com algumas transformações simples. A transformação logarítmica e a transformação de raiz quadrada são opções populares, especialmente quando a variância está mudando ao longo do tempo. Da mesma forma, remover uma tendência é geralmente feito por meio da diferenciação. Às vezes, uma série precisa ser diferenciada mais de uma vez. No entanto, se você se pegar fazendo diferenciação muitas vezes (mais de duas ou três), é improvável que consiga resolver o problema de estacionariedade apenas com diferenciação.

# Auto-correlação

**Entendendo e Identificando Autocorrelação**

De forma simples, a autocorrelação em uma série temporal é a ideia de que um valor em um determinado ponto no tempo pode estar correlacionado com o valor em outro ponto no tempo. Aqui, usamos o termo "autocorrelação" de maneira geral, sem entrar em detalhes técnicos.

Como exemplo, se você pegar uma série temporal anual de dados diários de temperatura, pode descobrir que comparar o dia 15 de maio de cada ano com o dia 15 de agosto de cada ano revela uma correlação. Isso significa que dias 15 de maio mais quentes tendem a estar associados a dias 15 de agosto também mais quentes (ou, alternativamente, mais frios). Você pode perceber que aprendeu algo interessante sobre o sistema de temperaturas, sugerindo que existe uma certa previsibilidade a longo prazo. Por outro lado, se a correlação for próxima de zero, isso também é interessante, pois indica que saber a temperatura do dia 15 de maio não fornece informações úteis sobre a temperatura esperada para o dia 15 de agosto.

Esse é um exemplo simples de autocorrelação.

A partir desse exemplo, expandiremos o conceito para a **função de autocorrelação**, que generaliza a autocorrelação ao considerar não apenas pontos específicos no tempo, mas qualquer dois pontos em uma série temporal, com uma distância fixa entre eles. Vamos explorar isso com mais detalhes, assim como a **autocorrelação parcial**.

A autocorrelação te dá uma ideia de como pontos de dados em diferentes momentos no tempo estão linearmente relacionados entre si, em função da diferença de tempo entre eles

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Gerar uma série temporal de exemplo (dados aleatórios com tendência)
np.random.seed(42)
y = np.random.normal(loc=0, scale=10, size=100) + np.arange(1, 101)

# Plotar a série temporal
plt.figure(figsize=(10, 4))
plt.plot(y)
plt.title("Série Temporal")
plt.show()

# Calcular e plotar a ACF
plt.figure(figsize=(10, 4))
plot_acf(y, lags=20)  # Calcula a ACF para os primeiros 20 lags
plt.show()
```
